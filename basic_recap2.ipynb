```python
import pandas as pd
import numpy as np
import matplotlib as plt

from sklearn.datasets import load_iris
```

```python
data = load_iris()
data
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names)
```

```python
df
```

```python
df.iloc[10:60:4,0] = np.nan
```

```python
df
```

```python
df.head(40)
```

```python
df.isnull().sum()
```

```python
x_missing = df[df.isnull().any(axis=1)]
x_nomiss = df.dropna()
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
x_train = x_nomiss.drop(columns=['sepal length (cm)'])
y_train = x_nomiss['sepal length (cm)']
x_test = x_missing.drop(columns = ['sepal length (cm)'])
regressor.fit(x_train, y_train)
```

```python
x_missing['sepal length (cm)'] = regressor.predict(x_test)
```

```python
df_final = pd.concat([x_missing, x_nomiss], axis=0)
df_final
```
