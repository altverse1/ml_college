```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('../Datasets/heart-disease .csv')
df
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(df.drop(columns=['target']),df['target'],test_size=0.25, random_state=4)
```

```python
gsc = GridSearchCV(RandomForestClassifier(),
    param_grid={
    'n_estimators':np.arange(70,140,5),
    'criterion':['gini', 'entropy', 'log_loss'],
    'max_depth':np.arange(10,50,5),
    'max_features':['sqrt', 'log2'],
    },
    cv=5,
    n_jobs=-1,
    )
gsc.fit(x_train, y_train)
```

```python
print('Best Estimator: ',gsc.best_estimator_)
print('Best Parameters: ',gsc.best_params_)
print('Best Score: ',gsc.best_score_)
```

```python
rsc = RandomizedSearchCV(RandomForestClassifier(), param_distributions=
                         {
                            'n_estimators':np.arange(70,140,5),
                            'criterion':['gini', 'entropy', 'log_loss'],
                            'max_depth':np.arange(10,50,5),
                            'min_samples_split':np.arange(1,20,2),
                            'max_features':['sqrt', 'log2'],
                         },
                         cv = 5,
                         scoring="r2",
                         verbose=1,
                         n_iter=50,
                         random_state=0,
                         n_jobs=-1 
                         )
rsc.fit(x_train,y_train)
```

```python
print('Best Estimator: ',rsc.best_estimator_)
print('Best Parameters: ',rsc.best_params_)
print('Best Score: ',rsc.best_score_)
```
