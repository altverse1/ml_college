```python
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get('https://news.google.com/rss?hl=en-US&gl=US&ceid=US:en')
response
```

```python
bs = BeautifulSoup(response.text, 'xml')
bs
```

```python
items = bs.find_all('item')
```

```python
from datetime import datetime, timedelta
import pandas as pd
cutoff = datetime.now()-timedelta(days=30)

data = []

for item in items:
    pubdate = datetime.strptime(item.pubDate.text, '%a, %d %b %Y %H:%M:%S %Z')
    if pubdate>cutoff:
        title = item.title.text
        link = item.link.text
        desc = item.description.text
        
        data.append([title,link,desc])

df = pd.DataFrame(data, columns=['Title','Link','Description'])
df
```

```python
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer

vec = TfidfVectorizer()
x = vec.fit_transform(df['Title'])
x
```

```python
memes = KMeans(n_clusters=5)
memes.fit(x)
```

```python
x_l= memes.labels_
x_l
```

```python
df['cluster'] = x_l
```

```python
df['cluster'].value_counts().sort_index()
```
