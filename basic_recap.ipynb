# California Dataset imputation
Trying out simple imputer

```python
from sklearn.datasets import fetch_california_housing
from sklearn.impute import SimpleImputer


imputer = SimpleImputer()

cali_data = fetch_california_housing()
```

```python
df_cali = pd.DataFrame(cali_data.data, columns=cali_data.feature_names)
df_cali
```

```python
#introducing NaN values
df_cali.iloc[::10] = np.nan
```

```python
df_cali.isnull().sum()
```

```python
df_imputed=imputer.fit_transform(df_cali)
df_calii = pd.DataFrame(df_imputed, columns=df_cali.columns) 
df_calii
```

# Missing Values Linear Regression

```python
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
```

```python
data = fetch_california_housing()
df = pd.DataFrame(data.data, columns=data.feature_names)
df
```

```python
df.iloc[10:20,0] = np.nan
df.head(11)
```

```python
x_missing = df[df.isna().any(axis=1)]
x_not_miss = df.dropna()
x_not_miss
```

```python
x_train = x_not_miss.drop(columns=['MedInc'])
y_train = x_not_miss['MedInc']
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
x_test = x_missing.drop(columns=['MedInc'])
```

```python
x_missing['MedInc'] = regressor.predict(x_test)
```

```python
df_final = pd.concat([x_missing,x_not_miss],axis = 0)
df_final
```
